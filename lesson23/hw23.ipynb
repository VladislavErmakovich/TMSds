Score: Оценка или рейтинг аниме

Popularity: Популярность аниме

Rank: Ранг аниме

Members: Количество членов, связанных с этим аниме (например, в сообществе или группе)

Description: Описание аниме

Synonyms: Синонимы или альтернативные названия для аниме (необязательное поле)

Japanese: Название на японском языке (необязательное поле)

English: Название на английском языке (необязательное поле)

Type: Тип аниме (например, TV, OVA, Movie)

Episodes: Количество эпизодов (необязательное поле)

Status: Статус аниме (например, Завершено, Онгоинг)

Aired: Дата первой и последней трансляции (необязательное поле)

Premiered: Дата премьеры (необязательное поле)

Broadcast: Детали трансляции (необязательное поле)

Producers: Производители аниме (например, студии)

Licensors: Держатели лицензии на аниме (необязательное поле)

Studios: Анимационные студии, создавшие аниме

Source: Источник материала (например, манга, ранобэ)

Genres: Жанры аниме (например, драма, фэнтези)

Demographic: Целевая аудитория (например, шонен, сейнен)

Duration: Продолжительность эпизода (необязательное поле)

Rating: Рейтинг контента (например, PG-13, R)
