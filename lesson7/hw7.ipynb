```python
"""

Создать класс Money для работы с денежными суммами. Число должно
быть представлено двумя полями: для рублей и для копеек. Копейки при
выводе на экран должна быть отделена от целой части запятой.
Реализовать сложение, вычитание, деление сумм, деление суммы на
дробное число, умножение на дробное число и операции сравнения.

"""

class Money():
    
    def __init__(self, rubles, cents): 
        self.rubles = rubles
        self.cents = cents
        if self.cents >= 100:
            self.rubles += self.cents // 100
            self.cents -= (self.cents // 100)*100
        
    def display_info(self):
        print(f"Балнс: {self.rubles},{self.cents}")
       
    def sum_money(self, rubles, cents):
        self.rubles += rubles
        self.cents += cents
        if self.cents >= 100:
            self.rubles += self.cents // 100
            self.cents -= (self.cents // 100)*100
            
    def subtraction_money(self, rubles, cents):
        if cents > self.cents:
            self.rubles -= 1
            self.cents += 100 - cents
            self.rubles -= rubles
        else:
            self.cents -= cents
            self.rubles -= rubles
        print(f"Балнс: {self.rubles},{self.cents}")
        
    def div_money(self, num):
        money = (self.rubles*100 + self.cents)/num
        self.rubles = int(money//100)
        self.cents =  int(money - (money//100)*100)
        print(f"Балнс: {self.rubles},{self.cents}")
        
    def mult_money(self, num):
        money = (self.rubles*100 + self.cents)*num
        self.rubles = int(money//100)
        self.cents =  int(money - (money//100)*100)
        print(f"Балнс: {self.rubles},{self.cents}")
        
    def compare(self, num):
        if self.rubles*100 + self.cents < num*100:
            print(f"У вас меньше денег")
        elif  self.rubles*100 + self.cents > num*100:
            print("У вас больше денег")
        else:
            print("У вас одинаковео количестов денег")
            
card = Money(25, 50)
card.display_info()
card.sum_money(4, 100)
card.subtraction_money(5, 55)
card.div_money(2.3)
card.mult_money(2.5)
card.compare(27.5)
```

```python
"""

Cоздать класс матрица и реализовать алгоритмы математических
операций над матрицами (сложение друг с другом, умножение на
скалярную величину).

"""

import random

class Matrix:
    
    def __init__(self, rows, colums):
        self.rows = rows
        self.colums = colums
        self.matrix = [[0 for j in range(self.colums)] for i in range(self.rows)]
        for i in range(self.rows):
            self.matrix[i] = list(map(lambda x: x + random.randint(1,9),self.matrix[i]))
                
    def display_matrix(self):
        print("\nМатрица:")
        for i in range(self.rows):
            print(self.matrix[i])
    
    def sum_matrix(self, matrix_sum):
        if self.rows != matrix_sum.rows or self.colums != matrix_sum.colums:
            return "У этих матриц не одинаковый размер"
        else:
            result = [[0 for j in range(self.colums)] for i in range(self.rows)]
            for i in range(self.rows):
                for j in range(self.colums):
                    result[i][j] = self.matrix[i][j] + matrix_sum.matrix[i][j]
        print("\nРезультат сложения двух матриц:")
        for i in range(self.rows):
            print(result[i])
    
    def mult_matrix(self, num):
        result = [[0 for j in range(self.colums)] for i in range(self.rows)]
        for i in range(self.rows):
            result[i] = list(map(lambda x: x*num, self.matrix[i]))
        print(f"\nРезультат умножения матрицы на число {num}:")
        for i in range(self.rows):
            print(result[i])
    
matrix_1 = Matrix(3, 2)
matrix_2 = Matrix(3, 2)
matrix_1.display_matrix()
matrix_2.display_matrix()
matrix_1.sum_matrix(matrix_2)
matrix_1.mult_matrix(2)
```
