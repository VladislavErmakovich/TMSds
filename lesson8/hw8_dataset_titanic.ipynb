```python
import pandas as pd
df =pd.read_csv("train.csv")
df.info()
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
#убираем все пустые значения в Age
df = df[df.Age.notna()]
df.shape
```

```python
#Сколько людей выжило и их возраст
df[["Age","Sex","Survived"]].groupby(["Sex"]).agg({"Age":["max","min","mean"],"Survived": "sum"})
```

```python
#Учет без экипажа
df_without_crew = df.query('Fare != 0')
df_without_crew.shape
```

```python
#Сколько пассажиров каждого класса выжило
df_without_crew[["Pclass","Survived"]].groupby(["Pclass"]).agg(["sum"])
```

```python
#Cтатистика по взрослым
adults = df_without_crew.loc[df_without_crew["Age"]>=18]
adults[["Sex","Survived","Age"]].groupby(["Sex"]).agg({"Age":["max","min","mean"], "Survived":["sum","count"]})
```

```python
#Cтатистика по детям
children= df_without_crew.loc[df_without_crew["Age"]<18]
children[["Sex","Survived","Age"]].groupby(["Sex"]).agg({"Age":["max","min","mean"], "Survived":["sum","count"]})
```

```python
df_without_crew[["Fare", "Survived", "Sex","Embarked"]].groupby(["Embarked","Sex","Survived"]).agg({"Fare" : ["count", "mean"]})
```

```python
persone_alone = df_without_crew.loc[(df_without_crew["SibSp"] == 0) & (df_without_crew["Parch"] == 0) ]
persone_alone.shape
```

```python
persone_alone[["Fare", "Survived","Embarked"]].groupby(["Embarked","Survived"]).agg({"Fare" : ["count", "mean"]})
```

```python
persone_alone[["Fare", "Survived"]].groupby(["Survived"]).agg({"Fare" : ["count", "mean"]})
```

```python
persone_with_family = df_without_crew.loc[(df_without_crew["SibSp"] != 0) | (df_without_crew["Parch"] != 0) ]
persone_with_family.shape
```

```python
persone_with_family[["Fare", "Survived","Embarked"]].groupby(["Embarked","Survived"]).agg({"Fare" : ["count", "mean"]})
```

```python
persone_with_family[["Fare", "Survived"]].groupby(["Survived"]).agg({"Fare" : ["count", "mean"]})
```
