```python
import numpy as np
import matplotlib.pyplot as plt

def inversion(picture: np.ndarray)-> None:
    """
    Inversion of one red channel
    
    :param first: picture input numpy.ndarray value
    :return: shows an image
    """
    print("Оригинал:")
    plt.imshow(picture)
    plt.show()
    
    inversion_red = 255 - picture[:,:,0]
    print("Инверсия канала red:")
    plt.imshow(inversion_red)
    plt.show()
    
    inversion_green = 255 - picture[:,:,1]
    print("Инверсия канала green:")
    plt.imshow(inversion_green)
    plt.show()
    
    inversion_blue = 255 - picture[:,:,2]
    print("Инверсия канала blue:")
    plt.imshow(inversion_blue)
    plt.show()
    
    inversion_all = 255 - picture
    print("Инверсия всех каналов(негатив):")
    plt.imshow(inversion_all)
    plt.show()

img_1 = plt.imread("image_1.jpg")
inversion(img_1)
img_2 = plt.imread("image.jpg")
inversion(img_2)
img_3 = plt.imread("image_3.jpg")
inversion(img_3)
```
