Сделаем оценку стран по качеству обучения и исследованиям

Но для начала избавимся от Unisted States of America, Unted Kingdom(такого не существует)

```python
error = df[df['country'] == 'Unisted States of America']
error
```

```python
error = df[df['country'] == 'Unted Kingdom']
error
```

```python
lst_unique_country = list(df.country.unique())

if 'Unisted States of America' in lst_unique_country:
  lst_unique_country.remove('Unisted States of America')

if 'Unted Kingdom' in lst_unique_country:
  lst_unique_country.remove('Unted Kingdom')
```

```python
lst_unique_country
```

```python
top_research = []
top_teaching = []

for i in lst_unique_country:
  x = df[df['country'] == i ]
  x_mean = x.research.mean()
  top_research.append(x_mean)

for i in lst_unique_country:
  x = df[df['country'] == i ]
  x_mean = x.teaching.mean()
  top_teaching.append(x_mean)
```

```python
df_top_research = pd.DataFrame({"country": lst_unique_country , "research": top_research })
sorted_df_top_research = df_top_research.reindex(df_top_research["research"].sort_values(ascending= True).index)
sorted_df_top_research
```

```python
df_top_teaching = pd.DataFrame({"country": lst_unique_country , "teaching": top_teaching })
sorted_df_top_teaching = df_top_teaching.reindex(df_top_teaching["teaching"].sort_values(ascending= True).index)
sorted_df_top_teaching
```

```python
plt.figure(figsize=(20,5))
sns.barplot(x = sorted_df_top_research['country'], y = sorted_df_top_research['research'],palette = 'pink')
plt.xlabel('County')
plt.ylabel('Research')
plt.xticks(rotation = 80)
plt.show()
```

```python
plt.figure(figsize=(20,5))
sns.barplot(x = sorted_df_top_teaching['country'], y = sorted_df_top_teaching['teaching'], palette='coolwarm')
plt.xlabel('County')
plt.ylabel('Teaching')
plt.xticks(rotation = 80)
plt.show()
```

```python
countries = []

for c in lst_unique_country:
  x = df[df['country']==c]
  x_cnt = x.country.count()
  countries.append(x_cnt)
```

```python
df_country = pd.DataFrame({"country" : lst_unique_country,"count": countries})
df_country_sort = df_country.reindex(df_country["count"].sort_values(ascending=True).index)
df_country_sort
```

```python
plt.figure(figsize=(20,5))
sns.barplot(x = df_country_sort["country"], y = df_country_sort["count"])
plt.xlabel('Страны')
plt.ylabel('Количество университетов')
plt.xticks(rotation = 80)
plt.show()
```

```python
#случайная(рандом), 1/4 значений от генеральной
df_selection_1 = df.sample(650)
df_selection_1.hist(figsize=(20, 10), color='blue', alpha = 0.4)
df_selection_1.describe()
```

```python
#брали значения с шагом 4
df_selection_2 = df[::4]
df_selection_2.hist(figsize=(20, 10), color='red',alpha = 0.8)
df_selection_2.describe()
```

```python
#брали значения из каждого года с шагом 4
df_selection_3 = pd.concat([df[::4] for df in [df_2011, df_2012, df_2013, df_2014, df_2015, df_2016]])
df_selection_3.hist(figsize=(20, 10), color='pink',alpha = 0.8)
df_selection_3.describe()
```
