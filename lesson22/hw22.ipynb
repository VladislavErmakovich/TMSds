```python
df_modified = df.copy()
df_modified = df_modified.drop(['Unnamed: 0.1','Unnamed: 0'],axis = 1)
```

**Перевод из категориальных в числовое**

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

categorial_colomns = []
for col in df_modified.columns:
  if df_modified[col].dtypes == 'object':
    categorial_colomns.append(col)

for col in categorial_colomns:
  df_modified.loc[:, col] = label_encoder.fit_transform(df_modified[col])
```

## **Регрессия**

```python
X_reg = df_modified.drop(['price'],axis = 1)
y_reg = df_modified['price']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier

from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score
```

**Построим для начала простые мдели без изменений(линейная регрессия, дерево решений)**

```python
X_train, X_test, y_train, y_test = train_test_split(X_reg, y_reg, test_size=0.25, random_state = 0)

#Линейная регрессия
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_train)
print(f'MAE(train): {mean_absolute_error(y_train, y_pred)}')

y_pred = model.predict(X_test)
print(f'MAE(test): {mean_absolute_error(y_test, y_pred)}')

y_pred = model.predict(X_reg)
print(f'MAE(test): {mean_absolute_error(y_reg, y_pred)}')
```

```python
#Дерево решений
model = DecisionTreeRegressor(criterion='squared_error',
                               max_depth = 20,
                               min_samples_split = 3,
                               min_samples_leaf=1,
                               random_state = 0)

model.fit(X_train, y_train)
y_pred = model.predict(X_train)
print(f'MAE(train): {mean_absolute_error(y_train, y_pred)}')

y_pred = model.predict(X_test)
print(f'MAE(test): {mean_absolute_error(y_test, y_pred)}')

y_pred = model.predict(X_reg)
print(f'MAE(test): {mean_absolute_error(y_reg, y_pred)}')
```

```python
#Случайный лес
model = RandomForestRegressor(n_estimators=75,
                              criterion='squared_error',
                              max_depth=20,
                              min_samples_split=5,
                              min_samples_leaf=1,
                              random_state = 0)
model.fit(X_train, y_train)
y_pred = model.predict(X_train)
print(f'MAE(train): {mean_absolute_error(y_train, y_pred)}')

y_pred = model.predict(X_test)
print(f'MAE(test): {mean_absolute_error(y_test, y_pred)}')

y_pred = model.predict(X_reg)
print(f'MAE(ALL): {mean_absolute_error(y_reg, y_pred)}')
```

Давольно большую ошибку показывает линейную регрессию, удовлетворительную показывает дерево решений, а случайный лес тоже большое значение ошибки -> с учетом, что у нас самое минимальное значение цены среди всех это 9999 индийских рупий

**Применим PCA для наших моделей**

```python
pca = PCA().fit(X_reg)
plt.figure(figsize=(8,6))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance vs. Number of Components')
plt.grid(True)
plt.show()
```

# **Классификация**

```python
from sklearn.metrics import classification_report
```

```python
X_cls = df_modified.drop(['brand'],axis = 1)
y_cls = df_modified['brand']

X_train, X_test, y_train, y_test = train_test_split(X_cls, y_cls, test_size=0.25, random_state = 0)
```

**PCA**

```python
pca = PCA().fit(X_cls)
plt.figure(figsize=(8,6))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance vs. Number of Components')
plt.grid(True)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_cls, y_cls, test_size=0.35, random_state=42)
pca = PCA(n_components=3)
X_train_transformed = pca.fit_transform(X_train)
X_test_transformed = pca.transform(X_test)
X_cls_transformed = pca.transform(X_cls)
```

```python
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.fit_transform(y_test)
y_cls_encoded = label_encoder.fit_transform(y_cls)
```

```python
#Дерево решений
model = DecisionTreeClassifier(criterion='gini',
                               max_depth=15,
                               min_samples_split=20,
                               min_samples_leaf=1,
                               random_state = 42)

model.fit(X_train_transformed, y_train_encoded)

y_pred = model.predict(X_train_transformed)
accuracy = accuracy_score(y_train_encoded, y_pred)
print(f'Accuracy(train) = {accuracy}')

y_pred = model.predict(X_test_transformed)
accuracy = accuracy_score(y_test_encoded, y_pred)
print(f'Accuracy(test) = {accuracy}')

y_pred = model.predict(X_cls_transformed)
accuracy = accuracy_score(y_cls_encoded, y_pred)
print(f'Accuracy(all) = {accuracy}')
```

```python
#Случайный лес
model = RandomForestClassifier(n_estimators=75,
                              criterion='gini',
                              max_depth=15,
                              min_samples_split=7,
                              min_samples_leaf=1,
                              random_state = 0)


model.fit(X_train_transformed, y_train_encoded)

y_pred = model.predict(X_train_transformed)
accuracy = accuracy_score(y_train_encoded, y_pred)
print(f'Accuracy(train) = {accuracy}')

y_pred = model.predict(X_test_transformed)
accuracy = accuracy_score(y_test_encoded, y_pred)
print(f'Accuracy(test) = {accuracy}')

y_pred = model.predict(X_cls_transformed)
accuracy = accuracy_score(y_cls_encoded, y_pred)
print(f'Accuracy(all) = {accuracy}')
```

```python
wcss = [] #сумма квадратов расстояний внутри кластера

for i in range(1, 14):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=13, random_state=0)
    kmeans.fit(X_cluster)
    wcss.append(kmeans.inertia_)

# Визуализация результатов
plt.plot(range(1, 14), wcss, marker = 'o')
plt.title('Метод "Локтя"')
plt.xlabel('Число кластеров')
plt.ylabel('WCSS')
plt.show()
```

исходя из графика будем брать 5 кластеров

```python
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=10)
kmeans_labels = kmeans.fit_predict(X_cluster)
labels = kmeans.labels_
```
