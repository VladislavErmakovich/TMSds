```python
import pandas as pd
df = pd.read_csv("train.csv")
df.info()
```

PassengerId — идентификатор пассажира


Survival — поле в котором указано спасся человек (1) или нет (0)


Pclass — содержит социально-экономический статус:
высокий
средний
низкий


Name — имя пассажира


Sex — пол пассажира

Age — возраст


SibSp — содержит информацию о количестве родственников 2-го порядка (муж, жена, братья, сетры)


Parch — содержит информацию о количестве родственников на борту 1-го порядка (мать, отец, дети)


Ticket — номер билета


Fare — цена билета


Cabin — каюта


Embarked — порт посадки:  C — Cherbourg, Q — Queenstown ,S — Southampton

```python
df.isnull().sum()
```

```python
#убираем значения, где встречаются null(по Age)
df = df[df.Age.notna()]
df.shape
```

```python
#убираем экипаж
df = df.query("Fare !=0")
df.shape
```

Так как число не сильно изменилось, так что в теории можно этого не делать, но мы продолжим анализ без учета экипажа

```python
#Данные по пассажирам(количетсво всех и выживших, возраст)
df[["Survived", "Age", "Sex"]].groupby(["Sex"]).agg({"Survived": ["count", "sum"],"Age": ["mean","max","min"]})
```

```python
#сколько выжило взрослых
df[["Survived", "Age", "Sex"]][df.Age >= 18].groupby(["Sex"]).agg({"Survived": ["count", "sum"],"Age": ["mean","max","min"]})
```

```python
#сколько выжило детей
df[["Survived", "Age", "Sex"]][df.Age < 18].groupby(["Sex"]).agg({"Survived": ["count", "sum"],"Age": ["mean","max","min"]})
```

```python
#разделим на части:выжившие/погибшие
df_survived = df.query("Survived == 1")
df_dead = df.query("Survived == 0")
print(df_survived.shape, df_dead.shape)
```

```python
#общая информация по выжившим
df_survived[["Survived", "Age", "Sex"]].groupby(["Sex"]).agg({"Age": ["count", "mean","max","min"]})
```

```python
#общая информация по погибшим
df_dead[["Survived", "Age", "Sex"]].groupby(["Sex"]).agg({"Survived": "count","Age": ["mean","max","min"]})
```

```python
#информация по взрослым(выжившие)
df_survived[["Survived", "Age", "Sex"]][df_survived.Age >=18].groupby(["Sex"]).agg({"Survived": "count","Age": ["mean","max","min"]})
```

```python
#информация по взрослым(погибшие)
df_dead[["Survived", "Age", "Sex"]][df_dead.Age >=18].groupby(["Sex"]).agg({"Survived": "count","Age": ["mean","max","min"]})
```

```python
#информация по детям(выжившие)
df_survived[["Survived", "Age", "Sex"]][df_survived.Age < 18].groupby(["Sex"]).agg({"Survived": "count","Age": ["mean","max","min"]})
```

```python
#информация по детям(погибшие)
df_dead[["Survived", "Age", "Sex"]][df_dead.Age < 18].groupby(["Sex"]).agg({"Survived": "count","Age": ["mean","max","min"]})
```

Можно сделать небольшой вывод, что смертность высока у мужчин, чем у женщин; выживаемость у детей больше -> так как приоритет отходил детям и женщинам, но у мальчиков показатель все равно меньше

```python
#каюты у вышивших
print(df_survived["Cabin"].notna().sum() / df_survived["PassengerId"].count() * 100)
```

```python
#каюты у погибших
print(df_dead["Cabin"].notna().sum() / df_dead["PassengerId"].count() * 100)
```

У людей, у кого были каюты, больше вероятность выживания, чем у людей без.
Сделаю акцент на том, что люди в момент трагедии находились не в каютах, а на палубе или в каких то зонах отдыхах. Другого объяснения, что у людей не было кают, я не могу найти



Порядок посадки:

Southampton -> Cherbourg -> Queenstown

```python
df_survived[["Fare","Embarked"]].groupby(["Embarked"]).agg(["mean"])
```

```python
df_dead[["Fare","Embarked"]].groupby(["Embarked"]).agg(["mean"])
```

Из двух верхних блоков можно предположить, что выживаемость больше у тех, кто отдал больше денег за билет(лучшее расположение кают). Отметим, что показатель по месту посадки Q одинаков(приблизительно).


```python
#сами(без какой либо родни) -> выжившие
print(df_survived[["Fare","Embarked"]][(df_survived.SibSp == 0) & (df_survived.Parch == 0)].groupby(["Embarked"]).agg({"Fare": ["mean","count"]}))
df_survived["PassengerId"][(df_survived.SibSp == 0) & (df_survived.Parch == 0)].count()
```

```python
#сами(без какой либо родни) -> погибшие
print(df_dead[["Fare","Embarked"]][(df_dead.SibSp == 0) & (df_dead.Parch == 0)].groupby(["Embarked"]).agg({"Fare": ["mean","count"]}))
df_dead["PassengerId"][(df_dead.SibSp == 0) & (df_dead.Parch == 0)].count()
```

```python
#с родней -> выжившие
print(df_survived[["Fare","Embarked"]][(df_survived.SibSp > 0) | (df_survived.Parch > 0)].groupby(["Embarked"]).agg({"Fare": ["mean","count"]}))
df_survived["PassengerId"][(df_survived.SibSp > 0) | (df_survived.Parch > 0)].count()
```

```python
#с родней -> погибшие
print(df_dead[["Fare","Embarked"]][(df_dead.SibSp > 0) | (df_dead.Parch > 0)].groupby(["Embarked"]).agg({"Fare": ["mean","count"]}))
df_dead["PassengerId"][(df_dead.SibSp > 0) | (df_dead.Parch > 0)].count()
```

Из четырех последних блоко можно сделать вывод, что выживаемость у тех, кто были одни меньше, чем у тех, кто был с родней. Отметим, что тут не хватает двух человек

```python
#информация по выжившим и их класс кают
df_survived[["Fare","Embarked","Pclass"]].groupby(["Embarked","Pclass"]).agg({"Fare": ["mean","count"]})
```

```python
#информация по погибшим и их класс кают
df_dead[["Fare","Embarked","Pclass"]].groupby(["Embarked","Pclass"]).agg({"Fare": ["mean","count"]})
```

```python
#Общее
print("Survived")
print(f"Class 1: {df_survived['PassengerId'][df_survived['Pclass'] == 1].count()}")
print(f"Class 2: {df_survived['PassengerId'][df_survived['Pclass'] == 2].count()}")
print(f"Class 3: {df_survived['PassengerId'][df_survived['Pclass'] == 3].count()}\n")

print("Dead")
print(f"Class 1: {df_dead['PassengerId'][df_dead['Pclass'] == 1].count()}")
print(f"Class 2: {df_dead['PassengerId'][df_dead['Pclass'] == 2].count()}")
print(f"Class 3: {df_dead['PassengerId'][df_dead['Pclass'] == 3].count()}\n")
```

```python
#сами(без какой либо родни) -> выжившие
df_survived[["Fare","Embarked","Pclass"]][(df_survived.SibSp == 0) & (df_survived.Parch == 0)].groupby(["Embarked","Pclass"]).agg({"Fare": ["mean","count"]})
```

```python
#сами(без какой либо родни) -> погибшие
df_dead[["Fare","Embarked","Pclass"]][(df_dead.SibSp == 0) & (df_dead.Parch == 0)].groupby(["Embarked","Pclass"]).agg({"Fare": ["mean","count"]})
```

```python
#с родней -> выжившие
df_survived[["Fare","Embarked","Pclass"]][(df_survived.SibSp > 0) & (df_survived.Parch > 0)].groupby(["Embarked","Pclass"]).agg({"Fare": ["mean","count"]})
```

```python
#с родней -> погибшие
df_dead[["Fare","Embarked","Pclass"]][(df_dead.SibSp > 0) & (df_dead.Parch > 0)].groupby(["Embarked","Pclass"]).agg({"Fare": ["mean","count"]})
```

```python
#По тем людям, кто был сам
print("Survived")
print(f"Class 1: {df_survived['PassengerId'][(df_survived['Pclass'] == 1) & ((df_survived.SibSp == 0) & (df_survived.Parch == 0))].count()}")
print(f"Class 2: {df_survived['PassengerId'][(df_survived['Pclass'] == 2) & ((df_survived.SibSp == 0) & (df_survived.Parch == 0))].count()}")
print(f"Class 3: {df_survived['PassengerId'][(df_survived['Pclass'] == 3) & ((df_survived.SibSp == 0) & (df_survived.Parch == 0))].count()}\n")

print("Dead")
print(f"Class 1: {df_dead['PassengerId'][(df_dead['Pclass'] == 1) & ((df_dead.SibSp == 0) & (df_dead.Parch == 0))].count()}")
print(f"Class 2: {df_dead['PassengerId'][(df_dead['Pclass'] == 2) & ((df_dead.SibSp == 0) & (df_dead.Parch == 0))].count()}")
print(f"Class 3: {df_dead['PassengerId'][(df_dead['Pclass'] == 3) & ((df_dead.SibSp == 0) & (df_dead.Parch == 0))].count()}\n")
```

```python
#По тем людям, кто был с родней
print("Survived")
print(f"Class 1: {df_survived['PassengerId'][(df_survived['Pclass'] == 1) & ((df_survived.SibSp > 0) & (df_survived.Parch > 0))].count()}")
print(f"Class 2: {df_survived['PassengerId'][(df_survived['Pclass'] == 2) & ((df_survived.SibSp > 0) & (df_survived.Parch > 0))].count()}")
print(f"Class 3: {df_survived['PassengerId'][(df_survived['Pclass'] == 3) & ((df_survived.SibSp > 0) & (df_survived.Parch > 0))].count()}\n")

print("Dead")
print(f"Class 1: {df_dead['PassengerId'][(df_dead['Pclass'] == 1) & ((df_dead.SibSp > 0) & (df_dead.Parch > 0))].count()}")
print(f"Class 2: {df_dead['PassengerId'][(df_dead['Pclass'] == 2) & ((df_dead.SibSp > 0) & (df_dead.Parch > 0))].count()}")
print(f"Class 3: {df_dead['PassengerId'][(df_dead['Pclass'] == 3) & ((df_dead.SibSp > 0) & (df_dead.Parch > 0))].count()}\n")
```

Из последних 6 блоков мы можем сделать вывод:

Общий: выживаемость выше у тех, кто покупал билеты дороже и находился в 1-2 классе(по большей части 1-ый класс), учитывая тот факт, что в спасательные шлюпки садили вначале 1 класс, а потом другие, то так оно и на самом деле.

Конкретее: Выживаемость у тех, кто был сам или с родней и еще при этом билеты были у них 1-го класса -> высокая. У 2-го класса такая ситуация, что у тех, кто был сам выживаемость меньше чем у тех, кто был с родными. Ситуация у 3-го класса совсем печальная, но выживаемость выше у тех, кто был с родней

Стоит отметить, что посадка выполнялась таким образом:

Southampton -> Cherbourg -> Queenstown
